```python
# 「戦争」ゲーム全コード

from random import shuffle

class Card:
    suits = ["spades", "hearts", "diamonds", "clubs"]
    
    values = [None, None,
             "2", "3", "4", "5", "6", "7", "8", "9",
             "10", "Jack", "Queen", "King", "Ace"]
    
    def __init__(self, v, s):
        """スート（マーク）も値も整数値です"""
        self.value = v
        self.suit  = s
    
    # カード比較演算子 "<" 用特殊メソッド    
    def __lt__(self, c2):
        if self.value < c2.value:
            return True
        if self.value == c2.value:
            if self.suit < c2.suit:
                return True
            else:
                return False
        return False
    
    # カード比較演算子 ">" 用特殊メソッド    
    def __gt__(self, c2):
        if self.value > c2.value:
            return True
        if self.value == c2.value:
            if self.suit > c2.suit:
                return True
            else:
                return False
        return False
    
    #Cardインスタンスのprint用特殊メソッド     
    def __repr__(self):
        v = self.values[self.value] + " of " + self.suits[self.suit]
        return v

# デッキ（カード一式となる52枚のトランプ）を表すクラス
class Deck:
    def __init__(self):
        self.cards = []
        for i in range(2,15):
            for j in range(4):
                self.cards.append(Card(i, j))
        #作成したトランプをシャッフル         
        shuffle(self.cards)

    #トランプからカードを1枚抜き取る為のメソッド         
    def rm_card(self):
        if len(self.cards) == 0:
            return
        return self.cards.pop()

# Playerクラスの定義
class Player:
    def __init__(self, name):
        self.wins = 0
        self.card = None
        self.name = name
    
# Gameクラスの定義
class Game:
    def __init__(self):

        #Player情報をセット
        name1 = input("pプレーヤー１の名前 ")
        name2 = input("pプレーヤー２の名前 ")
        self.p1 = Player(name1)
        self.p2 = Player(name2)
        
        #デッキ（トランプ ）を初期化
        self.deck = Deck()
     
    #ラウンド毎の勝敗表示メソッド     
    def wins(self, winner):
        w = "このラウンドは {}が勝ちました"
        w = w.format(winner)
        print(w)

    #カード引き出しメソッド     
    def draw(self, p1n, p1c, p2n, p2c):
        d = "{} は {}、{} は {} を引きました"
        d = d.format(p1n, p1c, p2n, p2c)
        print(d)
    
    # Game実行のメインメソッド    
    def play_game(self):
        cards = self.deck.cards
        print("戦争を始めます！")

        #カード（52枚）を使い果たすまでループ         
        while len(cards) >= 2:
            m = "q で終了、それ以外のキーでPlay:"
            response = input(m)
            if response == 'q':
                break
            p1c = self.deck.rm_card()
            p2c = self.deck.rm_card()
            p1n = self.p1.name
            p2n = self.p2.name
            self.draw(p1n, p1c, p2n, p2c)
            if p1c > p2c:
                self.p1.wins += 1
                self.wins(self.p1.name)
            else:
                self.p2.wins += 1
                self.wins(self.p2.name)
            print("{} : {} 勝 {} 敗".format(p1n, self.p1.wins, self.p2.wins ))
            print("{} : {} 勝 {} 敗".format(p2n, self.p2.wins, self.p1.wins ))
                
        win = self.winner(self.p1, self.p2)
        print("ゲーム終了、{} の勝利です！".format(win))

    #最終勝敗表示         
    def winner(self, p1, p2):
        if p1.wins > p2.wins:
            return p1.name
        if p1.wins < p2.wins:
            return p2.name
        return "引き分け！"
        
game = Game()
game.play_game()
```
