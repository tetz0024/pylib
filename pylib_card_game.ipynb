```python
class Card:
    suits = ["spades", "hearts", "diamonds", "clubs"]
    
    values = [None, None,
             "2", "3", "4", "5", "6", "7", "8", "9",
             "10", "Jack", "Queen", "King", "Ace"]
    
    def __init__(self, v, s):
        """スート（マーク）も値も整数値です"""
        self.value = v
        self.suit  = s
    
    # カード比較演算子 "<" 用特殊メソッド    
    def __lt__(self, c2):
        if self.value < c2.value:
            return True
        if self.value == c2.value:
            if self.suit < c2.suit:
                return True
            else:
                return False
        return False
    
    # カード比較演算子 ">" 用特殊メソッド    
    def __gt__(self, c2):
        if self.value > c2.value:
            return True
        if self.value == c2.value:
            if self.suit > c2.suit:
                return True
            else:
                return False
        return False
    
    #Cardインスタンスのprint用特殊メソッド     
    def __repr__(self):
        v = self.values[self.value] + " of " + self.suits[self.suit]
        return v

   
card1 = Card(10,2)
card2 = Card(11,3)
print(card1)
print(card2)
print(card1 > card2)
```

```python
from random import shuffle

class Card:
    suits = ["spades", "hearts", "diamonds", "clubs"]
    
    values = [None, None,
             "2", "3", "4", "5", "6", "7", "8", "9",
             "10", "Jack", "Queen", "King", "Ace"]
    
    def __init__(self, v, s):
        """スート（マーク）も値も整数値です"""
        self.value = v
        self.suit  = s
    
    # カード比較演算子 "<" 用特殊メソッド    
    def __lt__(self, c2):
        if self.value < c2.value:
            return True
        if self.value == c2.value:
            if self.suit < c2.suit:
                return True
            else:
                return False
        return False
    
    # カード比較演算子 ">" 用特殊メソッド    
    def __gt__(self, c2):
        if self.value > c2.value:
            return True
        if self.value == c2.value:
            if self.suit > c2.suit:
                return True
            else:
                return False
        return False
    
    #Cardインスタンスのprint用特殊メソッド     
    def __repr__(self):
        v = self.values[self.value] + " of " + self.suits[self.suit]
        return v

# デッキ（カード一式となる52枚のトランプ）を表すクラス
class Deck:
    def __init__(self):
        self.cards = []
        for i in range(2,15):
            for j in range(4):
                self.cards.append(Card(i, j))
        #作成したトランプをシャッフル         
        shuffle(self.cards)

    #トランプからカードを1枚抜き取る為のメソッド         
    def rm_card(self):
        if len(self.cards) == 0:
            return
        return self.cards.pop()

# カード一式（トランプ）を作成
deck1 = Deck()
# 作成したカードをprint
for card in deck1.cards:
    print(card)

# 抜き取ったカードをプリント
print(deck1.rm_card())
```
