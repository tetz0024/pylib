```python
import random

def hangman(word):
    wrong = 0
    stages = ["",
              "----------         ",
              "|                  ",
              "|        |         ",
              "|        0         ",
              "|      ( | )       ",
              "|       | |        ",
              "|                  "
             ]
    # ゲームで使用するwordをリスト化     
    rletters = list(word)
    # wordの文字数分を "_" でマスキング    
    board = ["_"] * len(word)
    win = False
    print("ハングマンへようこそ！")
    
    #ゲームの勝敗が決まるまで（wrongの値がstage-1を超えるまで=Hangmanの絵が完成されるまで）繰り返し     
    while wrong < len(stages) - 1:
        print("\n")
        msg = "1文字を予想してね"
        char = input(msg)
        #予想文字が正解文字に含まれていたら、正解箇所を予想文字で上書き         
        if char in rletters:
            cind = rletters.index(char)
            board[cind] = char
            #             
            rletters[cind] = '$'
        else:
            wrong += 1
        #現時点の正解状況をprint         
        print(" ".join(board))
        #hangmanをどこまで出力するか(wrongでスライスすると手前までになってしまうので+1)         
        e = wrong + 1
        #現時点のhangmanの出力状況         
        print("\n".join(stages[0:e]))
        if "_" not in board:
            print("あなたの勝ち！")
            print(" ".join(board))
            win = True
            break
    if not win:
        print("\n".join(stages[0:wrong+1]))
        print("あなたの負け！正解は{}".format(word))

word_list = ["dog","cat","bird","lion","horse","pig"]
hangman(random.choice(word_list))
```
